```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load and sort data
df = pd.read_csv('performance_data.csv')
df = df.sort_values(by='N')

# Extract NumPy arrays
N = df['N'].to_numpy()
time_ms = df['TimeConsumed(ms)'].to_numpy()
memory_kb = df['MemoryUsed(KB)'].to_numpy()
log_n = np.log(N)

# --- Plot 1: Time vs N with log(N) ---
plt.figure(figsize=(12, 6))
plt.plot(N, time_ms, label='Time Consumed (ms)', color='dodgerblue')
plt.plot(N, log_n, label='log(N)', linestyle='--', color='darkorange')
plt.xlabel('N')
plt.ylabel('Time (ms)')
plt.title('Time Consumed vs N (with log(N) reference)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('time_vs_n_with_logn.png')
plt.show()

# --- Plot 2: Memory vs N ---
plt.figure(figsize=(12, 6))
plt.plot(N, memory_kb, label='Memory Used (KB)', color='seagreen')
plt.xlabel('N')
plt.ylabel('Memory (KB)')
plt.title('Memory Usage vs N')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('memory_vs_n.png')
plt.show()
```
